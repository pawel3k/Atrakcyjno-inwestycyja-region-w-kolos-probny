```python
### ZAD 11

def funkcjaStozek(r,h):
  objetosc = 1/3*(math.pi)*(r**2)*h
  polePodstawy = (math.pi)*(r**2)
  l = math.sqrt((r**2)+(h**2))
  poleBoczne = (math.pi)*r*l
  return objetosc, polePodstawy, poleBoczne,l
   
```

```python
### ZAD 13

### a)

A = np.array([[3, 1, 2],
              [4, 5, 4],
              [3, 6, 4]])
np.linalg.det(A)
```

Wyznacznik = 1.99

```python
### b) (ABt)^-1*C

B = np.array([[3, 1, 2],
              [1, 0, 4],
              [2, -3, -2]])

C = np.array([[3, 1, 2],
              [1, 0, 4],
              [2, -3, -2]])

Bt = np.transpose(B) ## Transponujemy macierz

nawias = np.matmul(A, Bt) ### Liczymy pierwszy iloczyn w nawiasie

nawiasOdwr = np.linalg.inv(nawias) ### Odrwacamy macierz

wynik = np.matmul(nawiasOdwr, C) ## liczymy ostateczny wynik

wynik
```

```python
### c) At + B^-1 + 3C

At = np.transpose(A)
b_Odwr = np.linalg.inv(B)

wynik = At + b_Odwr + 3*C
wynik
```
